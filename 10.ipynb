```python
import matplotlib.pyplot as plt
import numpy as np

# Gerar dados aleatórios
np.random.seed(42)
dados1 = np.random.normal(0, 1, 100)
dados2 = np.random.normal(2, 1, 100)
dados3 = np.random.normal(-2, 1, 100)

# Combinar os dados
dados = [dados1, dados2, dados3]

# Criar a figura e os eixos
fig, ax = plt.subplots()

# Criar o gráfico de caixa e bigodes
ax.boxplot(dados)

# Definir os rótulos para o eixo x
ax.set_xticklabels(['Dados 1', 'Dados 2', 'Dados 3'])

# Definir o título e os rótulos para o gráfico
ax.set_title('Gráfico de Caixa e Bigodes')
ax.set_xlabel('Conjuntos de Dados')
ax.set_ylabel('Valores')

# Exibir o gráfico
plt.show()
```
